**  
1. What do REGION PROPOSALS entail?**

REGION PROPOSALS is a technique used in object detection to propose
candidate object locations in an image. These proposals can be generated
using various methods, such as selective search or edge boxes, and are
used as input to object detection models for further refinement and
classification.

**2. What do you mean by NON-MAXIMUM SUPPRESSION? (NMS)**

NON-MAXIMUM SUPPRESSION (NMS) is a post-processing step in object
detection that filters out overlapping object detections by selecting
only the highest scoring detection and suppressing others that have a
high degree of overlap with it.

**3. What exactly is mAP?**

mAP stands for mean Average Precision, which is a metric used to
evaluate the performance of object detection models. It takes into
account both precision and recall, and calculates the average precision
over a range of recall values.

**4. What is a frames per secondÂ (FPS)?**

Frames per second (FPS) is a measure of how many frames an object
detection model can process per second. It is an important metric for
real-time applications, such as autonomous driving or video
surveillance.

**5. What is an IOU (INTERSECTION OVER UNION)?**

IOU (INTERSECTION OVER UNION) is a metric used to evaluate the overlap
between two bounding boxes in object detection. It is defined as the
ratio of the area of overlap between two bounding boxes to the area of
their union.

**6. Describe the PRECISION-RECALL CURVE (PR CURVE)**

PRECISION-RECALL CURVE (PR CURVE) is a plot that shows the relationship
between precision and recall for different detection thresholds in an
object detection model. It is used to evaluate the performance of a
model and compare it to other models.

**7. What is the term "selective search"?**

"Selective search" is a popular region proposal algorithm used in object
detection that combines multiple low-level image features to generate
candidate object locations.

**8. Describe the R-CNN model's four components.**

The R-CNN model has four main components: (1) region proposal, which
generates candidate object locations, (2) feature extraction, which
extracts features from each candidate region, (3) object classification,
which uses the extracted features to classify each region as an object
or background, and (4) object localization, which refines the bounding
box coordinates of each detected object.

**9. What exactly is the Localization Module?**

The Localization Module is a component of the R-CNN model that refines
the bounding box coordinates of each detected object by predicting
offsets from the original bounding box coordinates generated by the
region proposal algorithm.

**10. What are the R-CNN DISADVANTAGES?**

Some disadvantages of the R-CNN model include its high computational and
memory requirements, as well as its slow processing speed. Additionally,
it requires training separate models for object detection and object
classification, which can be time-consuming and computationally
expensive.
